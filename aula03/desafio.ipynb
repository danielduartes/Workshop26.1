```python
# Bibliotecas a serem importadas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm # biblioteca que pega os valores convertidos com o 'matplotlib.colors' e retorna uma cor RGBA 
import matplotlib.colors as mcolors # biblioteca para normalização e conversão de cores 
```

```python
temperaturas = [] # lista vazia em que as temperaturas sorteadas vão ser armazenadas
dias = ['Domingo', 'Segunda-Feira', 'Terça-Feira', 'Quarta-Feira', 'Quinta-Feira', 'Sexta-Feira', 'Sábado'] # lista com os dias da semana
for i in range(0,7): # laço de 7 repetições, um para cada número a ser sorteado
    temperaturas.append(np.random.randint(20, 35)) # adicionando os números sorteados com 'numpy' na lista de temperaturas

# Printando as listas
print(temperaturas)
print(dias)
```

```python
# Transformando as listas em séries para criar o dataframe com os dados extraídos
series_temperaturas = pd.Series(temperaturas) 
series_dias = pd.Series(dias)
```

```python
df = pd.DataFrame() # criando o objeto do dataframe
df['temperatura'] = series_temperaturas # criando a coluna 'temperatura' com a série de temperaturas
df['dia'] = series_dias # criando a coluna 'dia' com a série dos dias da semana
df # visualizando o dataframe
```

```python
print(f'Temperatura média: {df['temperatura'].mean():.1f}°C') # pegando o valor médio das temperaturas do dataframe
print(f'Temperatura máxima: {df['temperatura'].max():.1f}°C') # pegando o valor máximo das temperaturas do dataframe
print(f'Temperatura mínima: {df['temperatura'].min():.1f}°C') # pegando o valor mínimo das temperaturas do dataframe
```

Gráfico em que é possível ver o crescimento ou decrescimento da temperatura com o passar dos dias

```python
plt.plot(df['dia'], df['temperatura']) # criando o gráfico de linhas com o método 'plot', tendo como parâmetros as colunas do dataframe 
plt.title('Temperatura ao longo da semana') # título do dataframe
plt.ylabel('Temperatura (°C)') # nome do eixo Y do gráfico
plt.xticks(rotation=45) # rotacionando os nomes do eixo X para evitar sobreposição entre eles
plt.tight_layout() # método utilizado para ajustar automaticamente o espaçamento interno da figura
plt.show() # plotando o gráfico
```

Gráfico utilizado para visualizar a variação de temperatura não só com valores numéricos, mas também através de cores

```python
norm = mcolors.Normalize(vmin=min(df['temperatura']), vmax=max(df['temperatura'])) # criação de um normalizador que vai mapear os valores de temperatura entre 0 e 1 com base no valor máximo e mínimo presente na coluna
cores = [cm.coolwarm(norm(t)) for t in df['temperatura']] # iterando os valores da coluna 'temperatura' e gerando uma cor para cada um

plt.figure(figsize=(9, 5)) # criando as dimensões de onde o gráfico vai ser plotado
plt.bar(df['dia'], df['temperatura'], color=cores) # definindo o tipo do gráfico, o que o eixo X e Y vão representar e as cores a serem utilizadas (as geradas pelo 'cm')
plt.title('Temperatura por dia da semana') # título do gráfico
plt.xlabel('Dias') # nome do eixo X
plt.ylabel('Temperatura (°C)') # nome do eixo Y
plt.xticks(rotation=45) #  rotacionando os nomes do eixo X para evitar sobreposição entre eles
plt.tight_layout() # método utilizado para ajustar automaticamente o espaçamento interno da figura
plt.show() # plotando o gráfico
```

Gráfico em que pode-se ver em quais dias da semana a temperatura média foi ultrapassada ou não 

```python
media = np.mean(df['temperatura']) # calculando a média das temperaturas usando o método 'mean' do 'numpy' e armazenando numa variável

plt.barh(df['dia'], df['temperatura'], color='steelblue') # definindo que o gráfico será um de barras horizontais, o que o eixo X e Y vão representar e a cor das barras
plt.axvline(media, color='red', linestyle='--', label=f'Média: {media:.1f}°C') # criando a linha pontilhada que irá representar a média das temperaturas
plt.title('Temperatura por dia da semana') # título do gráfico
plt.xlabel('Temperatura (°C)') # nome do eixo X
plt.legend() # criando a legenda, indicando o que a linha pontilhada significa
plt.tight_layout() # método utilizado para ajustar automaticamente o espaçamento interno da figura
plt.show() # plotando o gráfico
```
